```python
import nltk
nltk.download('stopwords')

from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

# Assuming `user_prompts` is your list of user prompts
user_prompts = [
    "This is the first user prompt.",
    "Here is the second user prompt.",
    "And another one for good measure."
]

vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'))

# This will convert your text data into a matrix of TF-IDF features
tfidf_matrix = vectorizer.fit_transform(user_prompts)
```
